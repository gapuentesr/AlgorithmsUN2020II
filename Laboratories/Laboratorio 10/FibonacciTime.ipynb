> <h2>Fibonacci Time</h2>

```
import matplotlib.pyplot as plt
import time
```

> <h2>Recursiva</h2>



```
def fibonacci_recursivo(n):
  if n==0 or n==1:
    return 1
  else:
    return fibonacci_recursivo(n-1)+fibonacci_recursivo(n-2)
```

> <h2>Iterada</h2>



```
def fibonacci_secuencial(n):
  actual=0
  menos1=1
  menos2=1
  if n==0 or n==1:
    actual=1
  else:
    for i in range(2,n+1):
      actual=menos1+menos2
      menos2=menos1
      menos1=actual

  return actual

start=time.time()
s=fibonacci_secuencial(500)
end=time.time()
print(s)
print(end-start)
```

> <h2>Análisis recursiva</h2>



```
x=[]
y1=[]
n=10
for i in range(n):
  start = time.time()
  fibonacci_recursivo(i)
  end = time.time()
  
  x.append(i)
  y1.append((end-start)*1000)

plt.plot(x,x,label='linear')
plt.plot(x,y1,label='Fibonacci recursivo')
plt.legend()
plt.xlim(0,n)
plt.ylim(0,n)
plt.xlabel("Número de datos")
plt.ylabel("Tiempo en milisegundos")
plt.title("Implementación recursiva de Fibonacci")
plt.show()
```

> <h2>Análisis Iterada</h2>

```
x=[]
y1=[]
n=500
for i in range(n):
  start = time.time()
  fibonacci_secuencial(i)
  end = time.time()
  x.append(i)
  y1.append((end-start)*1000)

plt.plot(x,x,label='Linear')
plt.xlim(0,n)
plt.ylim(0,n)
plt.plot(x,y1,label='Fibonacci secuencial')
plt.legend()

plt.xlabel("Número de datos")
plt.ylabel("Tiempo en milisegundos")
plt.title("Implementación iterada de Fibonacci")
plt.show()
```



> <h2>Comparativa entre iterada y recursiva</h2>



```
x=[]
y1=[]
y2=[]
n=35
for i in range(n):
  x.append(i)
  start = time.time()
  fibonacci_secuencial(i)
  end = time.time()
  y1.append((end-start)*1000)

  start = time.time()
  fibonacci_recursivo(i)
  end = time.time()
  y2.append((end-start)*1000)

plt.plot(x,x,label='linear')

plt.xlim(0,n)
plt.ylim(0,n)
plt.plot(x,y1,label='Fibonacci secuencial')
plt.plot(x,y2,label='Fibonacci recursivo')
plt.legend()

plt.xlabel("Número de datos")
plt.ylabel("Tiempo en milisegundos")
plt.title("Implementación iterada de Fibonacci")
plt.show()
```
