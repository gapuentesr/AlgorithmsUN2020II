<h1>The Fibonacci Numbers Algorithm with Tabulation</h1>

Revisit the Fibonacci numbers algorithm and implement it in a bottom-up manner.

<h1>Bottom-up dynamic programming</h1>

Notice how we started from Fib(6) and went on to break it into Fib(5) and Fib(4). We saw that these had not been evaluated so we broke them further until we hit the base case. Now letâ€™s look at an alternate approach, where we start from the base case, i.e., Fib(0) and Fib(1), and then build-up to the solution. We will keep storing all the subproblems along the way.

So, the plan is to start from Fib(0) and evaluate every Fibonacci number between 0 and n giving us the value of Fib(n).

Look at the coding playground below.

```
def fib(n):
  if n == 0:
    return 0
  if n == 1:
    return 1
  # table for tabulation
  table = [None] * (n+1) 
  table[0] = 0        # base case 1, fib(0) = 0
  table[1] = 1        # base case 2, fib(1) = 1
  # filling up tabulation table starting from 2 and going upto n
  for i in range(2,n+1):  
    # we have result of i-1 and i-2 available because these had been evaluated already
    table[i] = table[i-1] + table[i-2]  
  # return the value of n in tabulation table
  return table[n]    

print(fib(10000))
```
