<h1>Challenge 1: Find All Permutations of a String</h1>
Practice recursion skills

> **Problem statement**

Given a string, str, you are required to output an array containing all the possible permutations of that string. By permutation, we simply mean all the different arrangements of the characters of that string.

> **Input**

Your input will be a single string str.

str = "abc"

> **Output**

Your function should return a list of all possible permutations of the string str.

permutations(str) = ["abc", "acb", "bac", "bca", "cab", "cba"]

You should not change the prototype of function permutations, as this will affect testing. You can make another function of the desired prototype and call it from permutations.

> **Solution**


```
def permutations(str):
  if str == "": # base case
    return [""]
  permutes = []
  for char in str:
    subpermutes = permutations(str.replace(char, "", 1))    # recursive step
    for each in subpermutes:
      permutes.append(char+each)
  return permutes

def main():
  print (permutations("abc"))

main()
```
